```python
import matplotlib.pyplot as plt # Plotting library we'll use
import numpy as np # For array manipulation
print("Ready to study orbits!")
```

Data for semi-major axes in AU and orbital periods in years, for

`[Mercury, Venus, Earth, Mars, Ceres, Jupiter, Saturn, Halley's Comet, Uranus, Neptune, Pluto]`

```python
# Semi-major axis in astronomical units
semi_major_axes = np.array([0.39, 0.72, 1, 1.52, 2.77, 5.2, 9.54, 17.83, 19.19, 30.06, 39.48])

# Orbital period, in years
orbital_periods = np.array([0.24, 0.6, 1, 1.88, 4.6, 11.86, 29.46, 75.32, 84.01, 164.82, 284])
```

Plot $a$ (semi-major axis) as a function of $P$ (orbital period)

```python
plt.figure()
plt.scatter(semi_major_axes, orbital_periods, color = '#f5428d', s = 100, marker='*')
plt.ylabel('P (years)', fontsize=20)
plt.xlabel('a (AU)', fontsize=20)
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```
